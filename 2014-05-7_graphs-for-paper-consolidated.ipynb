```python
import attenuation_turns_converter

mickey=attenuation_turns_converter.attenuationConverter(datafile='2014-04-21_mickey_only.npz')
minnie=attenuation_turns_converter.attenuationConverter(datafile='2014-04-21_minnie_only.npz')

mickey_top=attenuation_turns_converter.attenuationConverter(datafile='2014-04-21_mickey_-15dbminnie.npz')
minnie_top=attenuation_turns_converter.attenuationConverter(datafile='2014-04-21_minnie_-15dbmickey.npz')

adjust_mickey = [item - mickey.db_list[9]+mickey_top.db_list[8] for item in mickey.db_list[9:]]
adjust_minnie = [item - minnie.db_list[22]+minnie_top.db_list[8] for item in minnie.db_list[22:]]

total_mickey=mickey_top.db_list[:8]+adjust_mickey
total_minnie=minnie_top.db_list[:8]+adjust_minnie

fig1=figure(1)
ax1=fig1.add_subplot(111)
ax1.set_ylabel('Attenuation (dB)')
ax1.set_xlabel('Micrometer Turns')
# ax4.set_title('Turns versus attenuation for mickey and minnie')
ax1.plot(total_mickey,'--o', total_minnie, '--o')
fig1.savefig('plots/attenuation_mickey_minnie.pdf')
```

```python
from scipy import interpolate

turns = range(12)

def total_turns_for_db(attenuation):
        atten_reduced=np.array(total)-attenuation
        freduced=interpolate.UnivariateSpline(turns, atten_reduced, s=0)
        return freduced.roots()
    
attenuation_list=[-5,-10,-15,-20,-25,-30,-35,-40, -45, -50, -55, -60, -70, -75]
for i in range(len(attenuation_list)):
    print ('For an attenuation of %d dB turn both knobs'
           ' %.3g times.'% (attenuation_list[i], total_turns_for_db(attenuation_list[i])))
```

```python
import interpolater

data_files=  [
            '2014-04-10_0_turns_both.npz',#'2014-04-10_4_turns_both.npz',
            '2014-04-10_6_turns_both.npz',
            '2014-04-10_8_turns_both.npz', '2014-04-10_10_turns_both.npz'
            ]

interp=interpolater.interpolater()
pow_list=range(len(data_files))
freq_list=range(len(data_files))
mult_list=range(len(data_files))
prefix='./data_for_processing/'

for i in range(len(data_files)):
    data=np.load(prefix+data_files[i])
    pow_list[i],freq_list[i],mult_list[i]=interp.convert(data['freq'],data['x'])
    # Interpolate to get power from voltages
    freq_list[i] = freq_list[i]/1e9
    # get frequency in GHz
    
fig3=figure(3)
ax3=fig3.add_subplot(1,1,1)
for i in range(len(data_files)):
    if i%3==0:
        color='grey'
    else:
        color='blue'
    ax3.plot(freq_list[i],pow_list[i],color=color)
ax3.set_yscale('log')
ax3.set_xlabel('Output Frequency (GHz)')
ax3.set_ylabel('Output Power (W)')
ax3.set_ylim([10**-11,10**-3])
ax3.set_xlim([freq_list[0][0], freq_list[0][-1]])
ax3.set_yticks([10**-4, 10**-5, 10**-6, 10**-7, 10**-8, 10**-9, 10**-10])


ax31 = ax3.twiny()
ax31.set_xlabel('Input Frequency (GHz)')
ax31.set_xlim([freq_list[0][0]/12.0, freq_list[0][-1]/12.0])
ax31.grid(False)
ax31.set_xticks([10, 11, 12, 13, 14])

fig3.savefig('plots/frequency_sweeps.pdf')
```
