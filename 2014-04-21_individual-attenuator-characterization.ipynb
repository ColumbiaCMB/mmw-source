```python
fig2=figure(2)
ax2=fig2.add_subplot(111)
ax2.set_ylabel('dBm')
ax2.set_xlabel('Turns')
ax2.set_title('Turns versus dBm for mickey and minnie')
ax2.plot(mickey.data['mickey_turns'],mickey_dbm,'--o',minnie.data['minnie_turns'][13:],minnie_dbm[13:],
     '--o')
```

```python
attenuation_list=[-5,-10,-15,-20,-25,-30,-35,-40]
for i in range(len(attenuation_list)):
    print ('For an attenuation of %d dB on mickey turn the knobs'
           ' %.3g times.'%(attenuation_list[i],mickey.give_turns(attenuation_list[i])))
```

```python
# give_turns automatically works with mickey_turns and I forgot to reset between trials, so I have to slice out the minnie data.
# Thus, this custom method is necessary.

from scipy import interpolate
def minnie_turns(attenuation):
        atten_reduced=np.array(minnie.db_list[13:])-attenuation
        freduced=interpolate.UnivariateSpline(minnie.data['minnie_turns'][13:],atten_reduced,s=0)
        return freduced.roots()
```

```python
attenuation_list=[-5,-10,-15,-20,-25,-30,-35,-40]
for i in range(len(attenuation_list)):
    print ('For an attenuation of %d dB on minnie turn the knobs'
           ' %.3g times.'%(attenuation_list[i],minnie_turns(attenuation_list[i])))
```

```python
mickey_top=attenuation_turns_converter.attenuationConverter(datafile='2014-04-21_mickey_-15dbminnie.npz')
minnie_top=attenuation_turns_converter.attenuationConverter(datafile='2014-04-21_minnie_-15dbmickey.npz')
```

This data was taken by setting one attenuator to -15 dB (the number of turns found above) and varying the other. The goal was to eliminate any effects we see from the saturation of the ZBD. However, it seems like the turns-attenuation relationship is really not linear.

```python
fig3=figure(3)
ax3=fig3.add_subplot(111)
ax3.set_ylabel('dB')
ax3.set_xlabel('Turns')
ax3.set_title('Turns versus attenuation for mickey and minnie, other attenuator at -15 dB')
ax3.plot(mickey_top.data['mickey_turns'],mickey_top.db_list,'--o',minnie_top.data['minnie_turns'],minnie_top.db_list,
     '--o')
```

```python
minnie_top_dbm=minnie_top.give_dbm(minnie_top.data['x'])
mickey_top_dbm=mickey_top.give_dbm(mickey_top.data['x'])
```
