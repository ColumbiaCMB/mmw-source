```python
import attenuation_turns_converter
```

```python
ac=attenuation_turns_converter.attenuationConverter()
```

```python
fig1=figure(1)
ax1=fig1.add_subplot(1,1,1)
ax1.set_xlabel('N turns')
ax1.set_ylabel('Attenuation (dB)')
ax1.plot(ac.data['mickey_turns'],ac.db_list,'--o')
```

The attenuation appears to remain constant until about 4 turns, then becomes linear until a floor around 9 turns. It is worth investigating the source further to see whether this is really the behavior of the attenuators or if it is the behavior of the detectors.

```python
fig2=figure(2)
ax2=fig2.add_subplot(1,1,1)
ax2.set_xlabel('N turns')
ax2.set_ylabel('Power (dBm)')
ax2.plot(ac.data['mickey_turns'],ac.give_dbm(ac.data['x']),'--o')
```

When we convert the voltage to power in dBm, we can see the linear section of the data occurs around -15 dBm. I contacted Virginia diodes, who responded that our detector saturated above -15 dBm (their slides documenting this effect are on the wiki). This suggests that the level section at low number of turns shows the behavior of the detector rather than the attenuator.

```python
attenuation_list=[-5,-10,-15,-20,-25,-30,-35,-40,-45,-50]
for i in range(len(attenuation_list)):
    print 'For an attenuation of %d dB turn the attenuator knobs %.3g times.'%(attenuation_list[i],ac.give_turns(attenuation_list[i])[0])
```

I wrote a method that takes a desired attenuation value and interpolates to give the number of turns to get that interpolation. This uses the data as recorded, so low levels of attenuation are probably wrong. Also, I cut off the noise floor when performing the attenuation, so the method will not give results much below -50 dB.

```python
noise=np.load('data_for_processing/2014-04-14_broadband-not-connected.npz')
dbm_noise=ac.give_dbm(noise['x'])
plot(ac.give_dbm(ac.data['x'][46:]),'--o')
plot(dbm_noise, '--o')
```

The green points is data taken from the lockin amplifier without anything hooked up (both the TTL out and input signal connected to their BNC cables, but nothing else). The blue points are the last few points from the attenuation measurements. Note that both data sets have been converted into dBm.

```python
fig4=figure(4)
ax41=fig4.add_subplot(2,1,2)
ax41.set_ylabel('dBm')
noisex=range(len(noise['x']))
ax41.set_title('Disconnected Data')
ax41.errorbar(noisex,noise['x'],yerr=noise['x_std'])


ax42=fig4.add_subplot(2,1,1,sharey=ax41)
ax42.set_ylabel('dBm')
datax=range(len(ac.data['x'][46:]))
ax42.set_title('Tail of Real Data')
ax42.errorbar(datax,ac.data['x'][46:],yerr=ac.data['x_std'][46:])
```

Plotting the raw voltage data convinces me that the 'floor' of the attenuation we observe is the noise floor from the lockin amplifier.
