```python
one_hour_drift = np.load('data_for_processing/2014-05-08_1hr_drift.npz')
stable = np.load('data_for_processing/2014-05-08_1hr_drift_stable.npz')
heated = np.load('data_for_processing/2014-05-08_1hr_drift_turn_heater_on.npz')
```

```python
import attenuation_turns_converter
ac=attenuation_turns_converter.attenuationConverter()

plot(one_hour_drift['time'], ac.give_dbm(one_hour_drift['x']))
```

```python
plot(stable['time'], stable['x'])
```

```python
stable_watts=np.array(ac.give_watts(stable['x']))
stable_dbm = ac.give_dbm(stable['x'])
```

```python
fig1=figure(1)
ax1=fig1.add_subplot(111)
ax1.set_ylabel('Fluctuation from\nmean output power (%)')
ax1.set_xlabel('Time (seconds)')
ax1.plot(stable['time'], 100 * (stable_watts/stable_watts.mean() - 1))

ylims1 = [-0.6, 0.6]
ax1.set_ylim(ylims1)

ax12=ax1.twinx()
ax12.set_ylabel('Output power ($\mu$ W)')
ax12.set_ylim([1e6 * (1+ylims1[0]/100.0) * stable_watts.mean(), 1e6 * (1+ylims1[1]/100.0)*stable_watts.mean()])
ax12.grid(False)

print 1e6 * stable_watts.mean()

fig1.savefig('plots/hour_drift_broadband.pdf', bbox_inches='tight')
```

```python
heated_real = heated['x'][len(stable['time']):]
heated_time = heated['time'][len(stable['time']):]

fig2=figure(2)
ax2=fig2.add_subplot(111)
ax2.set_ylabel('Fluctuation from\nmean output power (%)')
ax2.set_xlabel('Time (seconds)')
ax2.annotate('28 Celcius', xy=(400, 0.6), xytext=(1000, 0.6), arrowprops=dict(facecolor='black', shrink=0.05))
ax2.annotate('30 Celcius', xy=(1600,-0.1), xytext=(2500, -0.1), arrowprops=dict(facecolor='black', shrink=0.05))
heated_real_watts = np.array(ac.give_watts(heated_real))
ax2.plot(heated_time, 100*(heated_real_watts/heated_real_watts.mean() - 1))

ylims2 = [-0.4, 0.8]
ax2.set_ylim(ylims2)

ax22=ax2.twinx()
ax22.set_ylabel('Output power ($\mu$ W)')
ax22.set_ylim([1e6 * (1+ylims2[0]/100.0)*heated_real_watts.mean(), 1e6 * (1+ylims2[1]/100.0)*heated_real_watts.mean()])
ax22.grid(False)

fig2.savefig('plots/hour_drift_broadband_delta_t.pdf', bbox_inches='tight')
```

```python
single_freq_stable = np.load('data_for_processing/2014-05-15_1hr_drift_12 GHz.npz')

fig3=figure(2)
ax3=fig3.add_subplot(111)
ax3.set_ylabel('CW-mode fluctuation from\nmeanoutput power (%)')
ax3.set_xlabel('Time (seconds)')
single_freq_stable_watts = np.array(ac.give_watts(single_freq_stable['x'][831:]))
ax3.plot(single_freq_stable['time'][831:], 100*(single_freq_stable_watts/single_freq_stable_watts.mean() - 1))

ylims3 = [-2.5, 1.5]
ax2.set_ylim(ylims3)

ax32=ax3.twinx()
ax32.set_ylabel('Output power ($\mu$ W)')
ax32.set_ylim([1e6 * (1+ylims3[0]/100.0) * single_freq_stable_watts.mean(), 1e6 * (1+ylims3[1]/100.0) * single_freq_stable_watts.mean()])
ax32.grid(False)

print single_freq_stable_watts.mean()

fig3.savefig('plots/hour_drift_single_stable.pdf', bbox_inches='tight')
```

```python
single_freq_delta_t = np.load('data_for_processing/2014-05-15_1hr_drift_12_GHz_delta_t.npz')

fig4=figure(4)
ax4=fig4.add_subplot(111)
ax4.set_ylabel('CW-mode fluctuation from\nmean output power (%)')
ax4.set_xlabel('Time (seconds)')
ax4.annotate('28 Celcius', xy=(200, 2), xytext=(750, 2), arrowprops=dict(facecolor='black', shrink=0.05))
ax4.annotate('31 Celcius', xy=(1000, -0.5), xytext=(1500, -0.5), arrowprops=dict(facecolor='black', shrink=0.05))
single_freq_delta_t_watts = np.array(ac.give_watts(single_freq_delta_t['x'][2554:]))
ax4.plot(single_freq_delta_t['time'][2554:], 100*(single_freq_delta_t_watts/single_freq_delta_t_watts.mean() - 1))

ylims4 = [-1.5, 2.5]
ax4.set_ylim(ylims4)

ax42=ax4.twinx()
ax42.set_ylabel('Output power ($\mu$ W)')
ax42.set_ylim([1e6 * (1+ylims4[0]/100.0) * single_freq_delta_t_watts.mean(), 1e6 * (1+ylims4[1]/100.0) * single_freq_delta_t_watts.mean()])
ax42.grid(False)

fig4.savefig('plots/hour_drift_single_delta_t.pdf', bbox_inches='tight')
```
