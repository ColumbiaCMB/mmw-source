```python
import attenuation_turns_converter
ac=attenuation_turns_converter.attenuationConverter()
```

```python
stable_watts=ac.give_watts(stable['x'])
stable_dbm = ac.give_dbm(stable['x'])
```

```python
fig1=figure(1)
ax1=fig1.add_subplot(111)
ax1.set_ylabel('Broadband Output Power (dBm)')
ax1.set_xlabel('Time (seconds)')
ax1.plot(stable['time'], stable_dbm)

fig1.savefig('plots/hour_drift_broadband.pdf')
```

```python
heated_real = heated['x'][len(stable['time']):]
heated_time = heated['time'][len(stable['time']):]

fig2=figure(2)
ax2=fig2.add_subplot(111)
ax2.set_ylabel('Broadband Output Power (dBm)')
ax2.set_xlabel('Time (seconds)')
ax2.annotate('28 Celcius', xy=(300, -2.1e1-0.03), xytext=(1000, -2.1e1-0.025), arrowprops=dict(facecolor='black', shrink=0.05))
ax2.annotate('28 Celcius', xy=(2000, -2.1e1-0.065), xytext=(2500, -2.1e1-0.05), arrowprops=dict(facecolor='black', shrink=0.05))
ax2.plot(heated_time, ac.give_dbm(heated_real))
fig2.savefig('plots/hour_drift_broadband_delta_t.pdf')
```
