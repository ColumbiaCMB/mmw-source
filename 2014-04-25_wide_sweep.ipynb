```python
import interpolater

data_files=  [
            '2014-04-25_wide_sweep_0dB.npz', '2014-04-25_wide_sweep_-20dB.npz'
            ]

interp=interpolater.interpolater()
pow_list=range(len(data_files))
freq_list=range(len(data_files))
mult_list=range(len(data_files))
prefix='./data_for_processing/'

for i in range(len(data_files)):
    data=np.load(prefix+data_files[i])
    pow_list[i],freq_list[i],mult_list[i]=interp.convert(data['freq'],data['x'])
    freq_list[i] = freq_list[i]/1e9
    
fig=figure(1)
ax1=fig.add_subplot(1,1,1)
ax1.set_title('Power versus frequency at various attenuations')
for i in range(len(data_files)):
    ax1.plot(freq_list[i],pow_list[i])
#ax1.set_yscale('log')
ax1.set_xlabel('Frequency (GHz)')
ax1.set_ylabel('Power (W)')
```

```python
fig2=figure(2)
ax2=fig2.add_subplot(1,1,1)
ax2.set_title('Power versus frequency at various attenuations')
for i in range(len(data_files)):
    ax2.plot(freq_list[i],pow_list[i])
ax2.set_yscale('log')
ax2.set_xlabel('Frequency (GHz)')
ax2.set_ylabel('Power (W)')
```

```python
import interpolater

data_files=  [
            '2014-04-10_0_turns_both.npz',#'2014-04-10_4_turns_both.npz',
            '2014-04-10_6_turns_both.npz',
            '2014-04-10_8_turns_both.npz', '2014-04-10_10_turns_both.npz'
            ]

interp=interpolater.interpolater()
pow_list=range(len(data_files))
freq_list=range(len(data_files))
mult_list=range(len(data_files))
prefix='./data_for_processing/'

for i in range(len(data_files)):
    data=np.load(prefix+data_files[i])
    pow_list[i],freq_list[i],mult_list[i]=interp.convert(data['freq'],data['x'])
    # Interpolate to get power from voltages
    freq_list[i] = freq_list[i]/1e9
    # get frequency in GHz
    
fig3=figure(3)
ax3=fig3.add_subplot(1,1,1)
for i in range(len(data_files)):
    if i%3==0:
        color='grey'
    else:
        color='blue'
    ax3.plot(freq_list[i],pow_list[i],color=color)
ax3.set_yscale('log')
ax3.set_xlabel('Frequency (GHz)')
ax3.set_ylabel('Power (W)')
ax3.set_ylim([10**-11,10**-3])
ax3.set_xlim([freq_list[0][0], freq_list[0][-1]])
```
